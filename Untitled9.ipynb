სტატისტიკური მონაცემების გამოთვლა

````python
statistics = {
    'ConvertedCompYearly_mean': df['ConvertedCompYearly'].mean(),
    'ConvertedCompYearly_std': df['ConvertedCompYearly'].std(),
    'ConvertedCompYearly_median': df['ConvertedCompYearly'].median(),
    'ConvertedCompYearly_min': df['ConvertedCompYearly'].min(),
    'ConvertedCompYearly_max': df['ConvertedCompYearly'].max()
}
````

````python
for key, value in statistics.items():
    print(f"{key}: {value}")
````

````python
df['ConvertedCompYearly'].value_counts()
````

ხაზოვანი დიაგრამის შექმნა

````python
comp_yearly = df['ConvertedCompYearly']
mean = np.mean(comp_yearly)
median = np.median(comp_yearly)
std_dev = np.std(comp_yearly)

plt.figure(figsize=(10, 5))
plt.plot(comp_yearly, label='ConvertedCompYearly')
plt.axhline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axhline(median, color='g', linestyle=':', label=f'Median: {median:.2f}')
plt.fill_between(range(len(comp_yearly)), mean - std_dev, mean + std_dev, color='r', alpha=0.2, label=f'Standard Deviation: {std_dev:.2f}')
plt.legend()
plt.grid(True)
plt.title('Statistics of ConvertedCompYearly')
plt.show()
````
